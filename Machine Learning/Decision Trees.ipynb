```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from subprocess import check_call
import matplotlib.pyplot as plt
```

# Classification using Decision Trees

```python
#Creating conditions for the classification task
iris = load_iris()

X = iris.data[:, 2:] #Taking petal length and width as features
y = iris.target
```

```python
#Training Decision Tree Classifier for the dataset
tree_classifier = DecisionTreeClassifier(max_depth=2)
tree_classifier.fit(X, y)
```

```python
#Visualizing the trained Decision Tree
plot_tree(
    tree_classifier,
    feature_names = iris.feature_names[2:],
    class_names = iris.target_names,
    rounded = True,
    filled = True
)
```

```python
#Estimating class probabilites for the Classifier
tree_classifier.predict_proba([[5, 1.5]])
```

```python
#Predicting the class
tree_classifier.predict([[5, 1.5]])
```

# Regression using Decesion Trees

```python
#Training Decision Tree Regressor for the data set
tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X, y)
```

```python
#Visualizing tree structure for the regressor
plot_tree(tree_reg)
```
